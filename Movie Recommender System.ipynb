```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import os
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies = movies.drop(columns=['homepage','title','movie_id','status','production_countries'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df = 3,max_features=None,strip_accents='unicode',analyzer = 'word',token_pattern=r'\w{1,}',
           ngram_range=(1,3),
           stop_words='english')



movies['overview'] = movies['overview'].fillna(' ')
```

```python
tfv_matrix = tfv.fit_transform(movies['overview'])
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# computing sigmoid kernel
sigmoid = sigmoid_kernel(tfv_matrix,tfv_matrix)
```

```python
# reverse mapping of indices and movie titles

indices = pd.Series(movies.index,index=movies['original_title']).drop_duplicates
```

```python
def give_recommendation(original_title,sigmoid=sigmoid):
    #getting the index corresponding to original title
    
    idx = indices()[original_title]
    
    # get the pairwise similarity scores
    sig_scores = list(enumerate(sigmoid[idx]))
    
    
    #sorting movies
    sig_scores = sorted(sig_scores,key = lambda x:x[1],reverse=True)
    
    
    sig_scores = sig_scores[1:6]
    
    
    movies_indices = [i[0] for i in sig_scores]
    
    return movies['original_title'].iloc[movies_indices]
```

```python
give_recommendation('Apollo 18')
```
